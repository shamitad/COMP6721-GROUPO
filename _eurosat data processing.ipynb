```python
import os
import rasterio
import numpy as np
from skimage.transform import rescale
from scipy.ndimage import gaussian_filter
from skimage.io import imsave


def percent_stretch(img, lower_percent=2, upper_percent=98):
    lower_bound = np.percentile(img, lower_percent)
    upper_bound = np.percentile(img, upper_percent)
    img = np.clip(img, lower_bound, upper_bound)
    img = 255 * (img - lower_bound) / (upper_bound - lower_bound)
    return img.astype('uint8')


input_dir = '/content/AI datasets/EuroSAT/EuroSAT_MS'
output_dir = '/content/Processed_EuroSAT_MS'   
os.makedirs(output_dir, exist_ok=True)

for root, _, files in os.walk(input_dir):
    for file in files:
        if file.endswith('.tif'):
            input_path = os.path.join(root, file)
            
            rel_path = os.path.relpath(input_path, input_dir)
            output_subdir = os.path.join(output_dir, os.path.dirname(rel_path))
            os.makedirs(output_subdir, exist_ok=True)
            output_path = os.path.join(output_subdir, f"processed_{file.replace('.tif', '.png')}")

              with rasterio.open(input_path) as src:
                img = src.read([1, 2, 3]).astype('float32')   

                img = percent_stretch(img)
                img_smoothed = gaussian_filter(img, sigma=1)
                scale_factor = 3
                img_rescaled = rescale(img_smoothed, scale=(scale_factor, scale_factor), channel_axis=0, anti_aliasing=True)

                img_rescaled = (img_rescaled * 255 / img_rescaled.max()).astype('uint8')

                imsave(output_path, np.moveaxis(img_rescaled, 0, -1))
                print(f"Processed and saved: {output_path}")

print("Processing complete.")
```
